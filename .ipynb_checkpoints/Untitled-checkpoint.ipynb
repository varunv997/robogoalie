# Stereo Vision Overview

Stereo Vision refers to the perception of depth and 3-dimensional structure obtained on the basis of visual information obtained from two or more pinhole cameras. By comparing information about a scene from two vantage points, 3D information can be extracted by examination of the relative positions of objects in the two panels, similar to the human eyes.

Two cameras take pictures of the same scene, but they are separated by a distance â€“ exactly like our eyes. A computer compares the images while shifting the two images together over top of each other to find the parts that match. The shifted amount is called the disparity. The disparity at which objects in the image best match is used by the computer to calculate their distance.

For a human, the eyes change their angle according to the distance to the observed object. To a computer this represents significant extra complexity in the geometrical calculations (epipolar geometry). In fact the simplest geometrical case is when the camera image planes are on the same plane, which we will be considering in this paper.

Computer stereo vision is a part of the field of computer vision. It is sometimes used in mobile robotics to detect obstacles. Example applications include the ExoMars Rover and surgical robotics. 

# Stereo Geometry

## Description

When the images from both the cameras are superimposed, then the resulting image can be interpreted as above.

* x and x' are the apparent positions of the point of consideration with respect to camera O and O'.
* f is the focal length measured in pixels (process is described in an upcoming section).
* B is the base line, i.e the distance between the two cameras (measured in meters in our case).
* Z is the depth of the point from the baseline.

## Formula

\begin{align*}
depth = Z & = \frac{Bf}{x - x'} \\
\end{align*}

The above formula is obtained using the property of similar triangles.

Source: <a href = "https://en.wikipedia.org/wiki/Computer_stereo_vision#Detailed_definition">Wikipedia</a> and <a href = "http://docs.opencv.org/3.1.0/dd/d53/tutorial_py_depthmap.html">OpenCV Documentation</a>

# Calculation of focal length of a camera (in pixels)

Triangle Similarity is used to determine the focal length of a camera in pixels.

* An object of a known width, W (in centimeters in this case) is kept at a known distance D from the camera and several images are captured.
* The apparent width, P of the object in the image is calculated using imaging tools like paint or photoshop for all these images (in pixels).
* Arithmetic mean of all these values is taken to approximate the width.

### Formula

\begin{align*}
Focal\:Length = F & = \frac{PD}{W} \\
\end{align*}

### Parameters of test-camera

#### Device: HTC Desire 616 (Front Camera)

* The known object used was an A4 size sheet with a width W = 29.6 cm.
* The known distance, D = 150 cm.
* The apparent width (in pixels) for four test images : 278, 277, 279, 279.

#### Mean

```python
import numpy as np

#fixed distance
D = 150

#known width
W = 29.6

Parr = np.array([278, 277, 279, 279])
mean = np.mean(Parr)

#scaling the image by half
P = mean/2 
print (P)
```

#### Focal Length

```python
F = (P*D)/W

#round off to two digits
F = round(F, 2)

print (F)
```

Hence the focal length of the camera is obtained in pixels.

Source: <a href = "http://www.pyimagesearch.com/2015/01/19/find-distance-camera-objectmarker-using-python-opencv/">pyimagesearch.com</a> (OpenCV using python tutorials, Adrian Rosebrock)

# Stereo Vision Code

#### This is a new segment of code exclusive for stereo examination

```python
#import important packages
import cv2
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
#define constants
B = 0.6 #known distance between the two cameras (in meters)
F = 705.03 #known focal length of the cameras (in pixels)
```

```python
#define a display function for matplotlib
def disp(x, y):
    plt.subplot(1, 2, 1)
    plt.axis("off")
    plt.imshow(x)
    plt.title('Left Image')
    plt.subplot(1, 2, 2)
    plt.axis("off")
    plt.imshow(y)
    plt.title('Right Image')
    plt.show()
```

```python
#read images
left = cv2.imread('test1.jpg')
right = cv2.imread('test2.jpg')

#scale down the images by a factor of 2
resized1 = cv2.resize(left,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)
resized2 = cv2.resize(right,None,fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)

#display
disp(resized1, resized2)
```

```python
#convert the image to hsv
hsv1 = cv2.cvtColor(resized1, cv2.COLOR_BGR2HSV)
hsv2 = cv2.cvtColor(resized2, cv2.COLOR_BGR2HSV)

#display
disp(hsv1, hsv2)
```

```python
#mask colors
#calculated using a trackbar program
lower_blue = np.array([99, 159, 162])
upper_blue = np.array([107, 187, 255])

#apply the mask
mask1 = cv2.inRange(hsv1, lower_blue, upper_blue)
mask2 = cv2.inRange(hsv2, lower_blue, upper_blue)

#display
disp(mask1, mask2)
```

```python
#Apply erosion to remove noise
#erosion
kernel = np.ones((5,5), np.uint8)
eroded1 = cv2.erode(mask1, kernel, iterations = 1)
eroded2 = cv2.erode(mask2, kernel, iterations = 1)

#display
disp(eroded1, eroded2)
```

```python
#Apply dilation to exapnd the object back to original size
dilated1 = cv2.dilate(eroded1, kernel, iterations = 3)
dilated2 = cv2.dilate(eroded2, kernel, iterations = 3)

#display
disp(dilated1, dilated2)
```

```python
#finding contour
ret1,thresh1 = cv2.threshold(dilated1,127,255,0)
_, contours1, hierarchy1 = cv2.findContours(thresh1,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
ret2, thresh2 = cv2.threshold(dilated2, 127,255,0)
_, contours2, hierarchy2 = cv2.findContours(thresh2,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

#draw contour
cv2.drawContours(resized1, contours1, -1, (0,255,0), 3)
cv2.drawContours(resized2, contours2, -1, (0,255,0), 3)

#display
disp(resized1, resized2)
```

```python
#find the largest contour just in case:
lc1 = contours1[0]
lc2 = contours2[0]
for i in contours1:
	if (cv2.contourArea(lc1) > cv2.contourArea(i)):
		lc1 = i

for j in contours2:
	if (cv2.contourArea(lc2) > cv2.contourArea(j)):
		lc2 = j
        
#Find bounding rectangle parameters for the largest contour
rect1 = cv2.minAreaRect(lc1)
rect2 = cv2.minAreaRect(lc2)

#result
print(rect1, "\n", rect2)
```

```python
#take the mid pt
center1x = int(rect1[0][0])
center1y = int(rect1[0][1])
center2x = int(rect2[0][0])
center2y = int(rect2[0][1])

#define origin
originx = int(resized1.shape[1]/2)
originy = int(resized1.shape[0])
```

```python
#Draw Lines

#normal lines:
cv2.line(resized1, (originx,originy), (originx, 0), (0,0,255), 3)
cv2.line(resized2, (originx,originy), (originx, 0), (0,0,255), 3)

#draw lines to the center of the object
cv2.line(resized1, (originx,originy), (center1x, center1y), (0,0,0), 3)
cv2.line(resized2, (originx,originy), (center2x, center2y), (0,0,0), 3)

disp(resized1, resized2)
```

```python
#stereo calculation
disparity = abs(center1x-center2x) #disparity

#depth
z = (B*F)/disparity #dept
```

## Result

```python
#round off to two decimal places
z = round(z, 2)

#print the result
print(z)
```
